```python
# list of urls to scrape
urls = [
    "https://akshayakalpa.org/products/milk/",
    "https://akshayakalpa.org/products/paneer/",
    "https://akshayakalpa.org/products/akshayakalpa-multifloral-raw-honey",
    "https://akshayakalpa.org/products/greek-yoghurt/",
    "https://akshayakalpa.org/products/akshayakalpa-organic-cow-ghee",
    "https://akshayakalpa.org/products/akshayakalpa-organic-mango-drink",
    "https://akshayakalpa.org/products/akshayakalpa-organic-buttermilk-on-the-go",
    "https://akshayakalpa.org/products/country-eggs/",
    "https://akshayakalpa.org/products/akshayakalpa-organic-cheese",
    "https://akshayakalpa.org/products/curd/",
    "https://akshayakalpa.org/products/batter/",
    "https://akshayakalpa.org/vegetables/",
    "https://akshayakalpa.org/products/organic-milk/",
    "https://akshayakalpa.org/products/organic-buttermilk/",
    "https://akshayakalpa.org/products/coconut/",
    "https://akshayakalpa.org/products/bananas-yelakki/",
    "https://akshayakalpa.org/products/organic-butter/",
    "https://akshayakalpa.org/products/bread/",
    "https://akshayakalpa.org/products/akshayakalpa-organic-cold-pressed-virgin-coconut-oil",
    "https://akshayakalpa.org/blog/",
    "https://akshayakalpa.org/recipes/",
    "https://akshayakalpa.org/company/",
    "https://akshayakalpa.org/our-story/",
    "https://akshayakalpa.org/leadership/",
    "https://akshayakalpa.org/sustainability/",
    "https://akshayakalpa.org/privacy_policy.php",
    "https://akshayakalpa.org/research-and-development/",
    "https://akshayakalpa.org/faq/",
    "https://akshayakalpa.org/farm-visit-bangalore/",
    "https://akshayakalpa.org/farm-visit-chennai/",
    "https://akshayakalpa.org/farm-visit-hyderabad/"
]

# function to scrape data from a single url
def scrape_url(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        page_title = soup.title.string if soup.title else "No Title"
        text_content = " ".join([paragraph.get_text(strip=True) for paragraph in soup.find_all('p')])
        return {
            "page_title": page_title,
            "text_content": text_content
        }
    else:
        return {
            "page_title": "Error",
            "text_content": f"Failed to retrieve the webpage. Status code: {response.status_code}"
        }

# create an empty list to store the results
results = []

# iterate through the list of urls and scrape data
for url in urls:
    data = scrape_url(url)
    results.append(data)
import pandas as pd
# create a dataframe from the results
df = pd.DataFrame(results)

# save the dataframe to a csv file
df.to_csv('scraped_data.csv', index=False, encoding='utf-8')

print("Data has been saved to scraped_data.csv")
```
